#### Collecting Analytics Based on Site visits ####

```python
from faker import Faker
import random
events=10_000
pages=[
    "/", "/about", "/pricing", "/contact", "/blog", "/products", "/login", "/signup", "/dashboard", "/settings"
]
fake=Faker()
session_ids=[fake.uuid4() for _ in range(20)]
path="/api/event"
base_url="http://localhost:8002"
create_endpoint=f"{base_url}{path}"
refferers=[
    "https://www.google.com/", "https://www.facebook.com/", "https://www.twitter.com", "https://www.github.com/", "",
    "https://www.google.com/", "https://www.facebook.com/", "https://www.twitter.com", "https://www.github.com/", "https://www.linkedin.com/"
]

for i in range(events):
    page = random.choice(pages)
    user_agent=random.choice([
        fake.chrome,
        fake.firefox,
        fake.safari,
        fake.opera, 
        fake.user_agent
    ])()
    payload=dict(
        page=random.choice(pages),
        user_agent=user_agent,
        ip_address=fake.ipv4(),
        refferer=random.choice(refferers),
        session_id=random.choice(session_ids),
        duration=random.randint(5,300)
    )
    response = requests.post(create_endpoint, json=payload)
    if response.ok:
        data = response.json()
        print(data)
    else:
        print(response.text)
```
