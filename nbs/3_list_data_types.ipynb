```python
import requests
import pprint
path="/api/event"
base_url="http://localhost:8002"
endpoint=f"{base_url}{path}"
endpoint
```

```python
response=requests.get(endpoint)
print(response.ok)
```

```python
if response.ok:
    data=response.json()
    print(data)
```
